```python
from transformers import AutoProcessor, AutoModel
from IPython.display import Audio
import torchaudio
import os
from pydub import AudioSegment
```

```python
processor = AutoProcessor.from_pretrained("suno/bark-small")
```

```python
model = AutoModel.from_pretrained("suno/bark-small")
```

```python
inputs = processor(
    text=["こんにちは、私の名前はクリッシーです"],
    return_tensors="pt",
)
```

```python
print(inputs)
```

```python
speech_values = model.generate(**inputs, do_sample=True)
```

```python
from IPython.display import Audio

sampling_rate = model.generation_config.sample_rate
Audio(speech_values.cpu().numpy().squeeze(), rate=sampling_rate)
```

```python

# Specify the path to the output MP3 audio file
output_audio_file = "./output_audio.mp3"  # Specify the desired output MP3 audio file path

# Create the output directory if it doesn't exist
output_dir = os.path.dirname(output_audio_file)
os.makedirs(output_dir, exist_ok=True)

# Convert the audio to a format compatible with pydub (e.g., int16)
audio_data = speech_values.cpu().numpy().squeeze().astype("int16")

# Create an AudioSegment from the audio data
audio_segment = AudioSegment(
    audio_data.tobytes(),
    frame_rate=sampling_rate,
    sample_width=audio_data.dtype.itemsize,
    channels=1  # Mono audio
)

# Export the AudioSegment as MP3
audio_segment.export(output_audio_file, format="mp3")

# Display the audio
Audio(speech_values.cpu().numpy().squeeze(), rate=sampling_rate)
```
